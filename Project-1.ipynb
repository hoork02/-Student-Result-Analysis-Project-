```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info() ##data type not null count ##
```

```python
df.isnull().sum()
```

```python
#drop unname column##
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
#change weekly hours
```

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("5-Oct","5-10")
df.head()
```

```python
##gender distribution##
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
## females are more than males in the data as we analyse the data ##
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot=True)
plt.title("Relation between student score and parent education")
plt.show()
```

```python
## education of parents have a good impact on children's education ##
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot=True)
plt.title("Relation between student score and parent marital statuas")
plt.show()
```

```python
## no/negliible impact on the student score related to marital stauts of parents##
```
